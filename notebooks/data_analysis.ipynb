# Credit Scoring — Data Analysis
Exploratory analysis of `dataset/credit_data.csv`, mirroring the EDA performed in `train_model.py`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
df = pd.read_csv('../dataset/credit_data.csv')
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print('Duplicate rows:', df.duplicated().sum())
```

```python
df.describe()
```

```python
df['creditworthy'].value_counts(normalize=True).plot(kind='bar', color=['#3FB68B', '#E5484D'])
plt.title('Class Distribution')
plt.show()
```

```python
numeric_cols = ['age', 'income', 'loan_amount', 'loan_duration_months', 'debt_ratio', 'payment_history_score']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.histplot(df[col].dropna(), kde=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
corr = df[numeric_cols + ['creditworthy']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(data=df, x='creditworthy', y='income')
plt.title('Income vs Credit Risk')
plt.show()
```
